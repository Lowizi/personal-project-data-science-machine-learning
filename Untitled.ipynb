```python
import pandas as pd

# Load the Excel sheet
original = pd.read_excel(
    "TEHD.xlsx",
    sheet_name="TEHD"
)

# Load the sample CSV
sample = pd.read_csv("ptad_nhis_dtad.csv")
```

```python
print(original.columns.tolist())
```

```python
print(sample.columns.tolist())
```

```python


column_map = {
    "pension_id": "PensionerID",
    "pension_number": "PenNumber",
    "pension_unique_no": "PensionerUniqueNo",
    "bvn": "BVN",
    "first_name": "FirstName",
    "last_name": "LastName",
    "middle_name": "MiddleName",
    "date_of_birth": "BirthDate",
    "gender": "Sex",
    "phone_number": "PhoneNumber.1",
    "home_address": "Address",
    "local_government_area": "LGA",
    "bank_name": "Bank",
    "account_number": "AccountNumber ON DB",
    "department": "Department",
    "mda": "MDA",
    "date_of_retirement": "RetirementDate",
    "state_of_origin":"State"
}
output = pd.DataFrame()

for col in sample.columns:

    # Special handling for account_name
    if col == "account_name":
        output[col] = (
            original["FirstName"].fillna("").astype(str).str.strip()
            + " "
            + original["LastName"].fillna("").astype(str).str.strip()
        ).str.strip()

    # Columns that exist in the source
    elif col in column_map:
        output[col] = original[column_map[col]]

    # Columns not available in source
    else:
        output[col] = ""

# Save using exact column order from sample file
output = output[sample.columns]

output.to_csv("supporting_documents.csv", index=False)

print("supporting_documents.csv created successfully")
print(f"Rows: {len(output)}")
print(f"Columns: {len(output.columns)}")
```
