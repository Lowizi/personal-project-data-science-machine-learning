```python
import pandas as pd

# Load datasets
properties = pd.read_csv("properties.csv")
metrics = pd.read_csv("metrics.csv")

# Merge datasets
df = properties.merge(
    metrics,
    on="Property ID",
    how="inner"
)

# Check data
print(df.head())
print(df.shape)
```

```python
df["Annual Gross Revenue (£)"] = (
    df["Monthly Rent (Ł)"] * 12
)
```

```python
df["Monthly Rent (Ł)"] = (
    df["Monthly Rent (Ł)"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
df["Maintenance Cost (£)"] = (
    df["Maintenance Cost (£)"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
print(df.dtypes)
```

```python
df["Annual Gross Revenue"] = df["Monthly Rent (Ł)"] * 12

df["Occupied Revenue"] = (
    df["Monthly Rent (Ł)"]
    * df["Occupied Months (2024)"]
)

df["Annual Void Cost"] = (
    df["Annual Gross Revenue"]
    - df["Occupied Revenue"]
)
```

```python
def extract_city(location):
    if "London" in location:
        return "London"
    elif "Manchester" in location:
        return "Manchester"
    elif "Leeds" in location:
        return "Leeds"
    elif "Edinburgh" in location:
        return "Edinburgh"
    elif "Birmingham" in location:
        return "Birmingham"
    elif "Bristol" in location:
        return "Bristol"
    elif "Glasgow" in location:
        return "Glasgow"
    return "Other"

df["City"] = df["Location"].apply(extract_city)
```

```python
df["Norm_Maintenance_Ratio"] = (
    (df["Maintenance Cost Ratio (%)"] - 0.45)
    / (6.67 - 0.45)
)

df["Norm_Void_Cost"] = (
    df["Annual Void Cost"]
    / 18600
)

df["Norm_Satisfaction_Risk"] = 1 - (
    (df[satisfaction_col] - 5.4)
    / (9.5 - 5.4)
)

df[
    [
        "Property ID",
        "Norm_Maintenance_Ratio",
        "Norm_Void_Cost",
        "Norm_Satisfaction_Risk"
    ]
].head()
```

```python
df["Composite Risk Score"] = (
    0.30 * df["Norm_Maintenance_Ratio"]
    + 0.40 * df["Norm_Void_Cost"]
    + 0.30 * df["Norm_Satisfaction_Risk"]
)

df["Composite Risk Score"] = (
    df["Composite Risk Score"]
    .round(4)
)
```

```python
def risk_tier(score):
    if score >= 0.60:
        return "High Risk"
    elif score >= 0.30:
        return "Medium Risk"
    else:
        return "Low Risk"

df["Risk Tier"] = df["Composite Risk Score"].apply(risk_tier)
```

```python
import pandas as pd

from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from openpyxl.chart import BarChart, ScatterChart, Reference, Series

# -----------------------------
# SAVE DATAFRAMES
# -----------------------------

with pd.ExcelWriter(
    "NestMetrics_Analysis.xlsx",
    engine="openpyxl"
) as writer:

    df.to_excel(
        writer,
        sheet_name="Main_Data",
        index=False
    )

    pivot.to_excel(
        writer,
        sheet_name="E10_Pivot",
        index=False
    )

    # Dashboard KPIs
    dashboard = pd.DataFrame({
        "KPI": [
            "Total Annual Portfolio Revenue",
            "Total Annual Void Cost",
            "Portfolio Average Composite Risk Score",
            "Number of High Risk Properties",
            "Total Revenue Recovery Potential"
        ],
        "Value": [
            590160,
            73530,
            0.2395,
            2,
            44730
        ]
    })

    dashboard.to_excel(
        writer,
        sheet_name="Dashboard",
        index=False
    )

# -----------------------------
# OPEN WORKBOOK
# -----------------------------

wb = load_workbook(
    "NestMetrics_Analysis.xlsx"
)

ws = wb["Main_Data"]

# -----------------------------
# Q7 RED HIGHLIGHT
# -----------------------------

red_fill = PatternFill(
    start_color="FF9999",
    end_color="FF9999",
    fill_type="solid"
)

ratio_col = None

for cell in ws[1]:
    if cell.value == "Maintenance Cost Ratio (%)":
        ratio_col = cell.column
        break

for row in range(2, ws.max_row + 1):

    value = ws.cell(
        row=row,
        column=ratio_col
    ).value

    if value > 3:
        ws.cell(
            row=row,
            column=ratio_col
        ).fill = red_fill

# -----------------------------
# Q9 ORANGE HIGHLIGHT
# -----------------------------

orange_fill = PatternFill(
    start_color="FFD580",
    end_color="FFD580",
    fill_type="solid"
)

risk_col = None

for cell in ws[1]:
    if cell.value == "Risk Flag":
        risk_col = cell.column
        break

for row in range(2, ws.max_row + 1):

    value = ws.cell(
        row=row,
        column=risk_col
    ).value

    if value == "AT RISK":
        ws.cell(
            row=row,
            column=risk_col
        ).fill = orange_fill

# -----------------------------
# CHART SHEET
# -----------------------------

chart_sheet = wb.create_sheet(
    "Charts"
)

# Copy city yield table

for row in range(len(city_yield)):

    chart_sheet.cell(
        row=row + 2,
        column=1,
        value=city_yield.iloc[row]["City"]
    )

    chart_sheet.cell(
        row=row + 2,
        column=2,
        value=city_yield.iloc[row]["Net Annual Yield"]
    )

chart_sheet["A1"] = "City"
chart_sheet["B1"] = "Net Annual Yield"

# -----------------------------
# BAR CHART
# -----------------------------

bar = BarChart()

bar.title = "Net Annual Yield by City"

data = Reference(
    chart_sheet,
    min_col=2,
    min_row=1,
    max_row=len(city_yield) + 1
)

cats = Reference(
    chart_sheet,
    min_col=1,
    min_row=2,
    max_row=len(city_yield) + 1
)

bar.add_data(
    data,
    titles_from_data=True
)

bar.set_categories(cats)

chart_sheet.add_chart(
    bar,
    "E2"
)

# -----------------------------
# SCATTER CHART
# -----------------------------

scatter = ScatterChart()

scatter.title = (
    "Risk Score vs Net Annual Yield"
)

scatter.x_axis.title = (
    "Composite Risk Score"
)

scatter.y_axis.title = (
    "Net Annual Yield"
)

# Put scatter data

chart_sheet["H1"] = "Composite Risk Score"
chart_sheet["I1"] = "Net Annual Yield"

for i, row in enumerate(df.itertuples(), start=2):

    chart_sheet.cell(
        row=i,
        column=8,
        value=row._asdict()["Composite Risk Score"]
    )

    chart_sheet.cell(
        row=i,
        column=9,
        value=row._asdict()["Net Annual Yield"]
    )

xvalues = Reference(
    chart_sheet,
    min_col=8,
    min_row=2,
    max_row=len(df) + 1
)

yvalues = Reference(
    chart_sheet,
    min_col=9,
    min_row=2,
    max_row=len(df) + 1
)

series = Series(
    yvalues,
    xvalues
)

scatter.series.append(series)

chart_sheet.add_chart(
    scatter,
    "E20"
)

# -----------------------------
# SAVE
# -----------------------------

wb.save(
    "NestMetrics_Analysis.xlsx"
)

print(
    "NestMetrics_Analysis.xlsx created successfully!"
)
```

```python
import pandas as pd

from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from openpyxl.chart import BarChart, ScatterChart, Series, Reference

# ==========================
# EXPORT DATA
# ==========================

with pd.ExcelWriter(
    "NestMetrics_Analysis.xlsx",
    engine="openpyxl"
) as writer:

    # Main sheet
    df.to_excel(
        writer,
        sheet_name="Main_Data",
        index=False
    )

    # Pivot sheet
    pivot.to_excel(
        writer,
        sheet_name="E10_Pivot",
        index=False
    )

# ==========================
# LOAD WORKBOOK
# ==========================

wb = load_workbook(
    "NestMetrics_Analysis.xlsx"
)

ws = wb["Main_Data"]

# ==========================
# Q7 RED FORMATTING
# ==========================

red_fill = PatternFill(
    start_color="FF0000",
    end_color="FF0000",
    fill_type="solid"
)

ratio_col = list(df.columns).index(
    "Maintenance Cost Ratio (%)"
) + 1

for row in range(2, len(df) + 2):

    cell = ws.cell(
        row=row,
        column=ratio_col
    )

    if cell.value > 3:
        cell.fill = red_fill

# ==========================
# Q9 ORANGE FORMATTING
# ==========================

orange_fill = PatternFill(
    start_color="FFA500",
    end_color="FFA500",
    fill_type="solid"
)

risk_col = list(df.columns).index(
    "Risk Flag"
) + 1

for row in range(2, len(df) + 2):

    cell = ws.cell(
        row=row,
        column=risk_col
    )

    if cell.value == "AT RISK":
        cell.fill = orange_fill

# ==========================
# DASHBOARD SHEET
# ==========================

dashboard = wb.create_sheet(
    "Dashboard"
)

dashboard["A1"] = "NestMetrics Executive Dashboard"

dashboard["A3"] = "Total Revenue"
dashboard["B3"] = df["Annual Gross Revenue"].sum()

dashboard["A4"] = "Total Void Cost"
dashboard["B4"] = df["Annual Void Cost"].sum()

dashboard["A5"] = "Average Risk Score"
dashboard["B5"] = round(
    df["Composite Risk Score"].mean(),
    4
)

dashboard["A6"] = "High Risk Properties"
dashboard["B6"] = (
    df["Risk Tier"] == "High Risk"
).sum()

dashboard["A7"] = "Recovery Potential"
dashboard["B7"] = df[
    "Revenue Recovery Potential"
].sum()

# ==========================
# BAR CHART DATA
# ==========================

dashboard["E1"] = "City"
dashboard["F1"] = "Net Yield"

city_yield = (
    df.groupby("City")["Net Annual Yield"]
    .sum()
    .reset_index()
)

for i, row in enumerate(
    city_yield.itertuples(),
    start=2
):

    dashboard.cell(
        row=i,
        column=5,
        value=row.City
    )

    dashboard.cell(
        row=i,
        column=6,
        value=row._2
    )

# ==========================
# BAR CHART
# ==========================

bar = BarChart()

bar.title = "Net Annual Yield by City"

bar.x_axis.title = "City"
bar.y_axis.title = "Net Annual Yield (£)"

bar.add_data(
    data,
    titles_from_data=True
)

bar.set_categories(cats)

dashboard.add_chart(
    bar,
    "H2"
)
# ==========================
# SCATTER DATA SHEET
# ==========================

scatter_sheet = wb.create_sheet(
    "Scatter_Data"
)

scatter_sheet["A1"] = "Property ID"
scatter_sheet["B1"] = "Composite Risk Score"
scatter_sheet["C1"] = "Net Annual Yield"

for i, (_, row) in enumerate(
    df.iterrows(),
    start=2
):

    scatter_sheet.cell(
        row=i,
        column=1,
        value=row["Property ID"]
    )

    scatter_sheet.cell(
        row=i,
        column=2,
        value=row["Composite Risk Score"]
    )

    scatter_sheet.cell(
        row=i,
        column=3,
        value=row["Net Annual Yield"]
    )

# ==========================
# SCATTER CHART
# ==========================
scatter = ScatterChart()

scatter.title = "Risk Score vs Net Yield"

scatter.x_axis.title = "Composite Risk Score"
scatter.y_axis.title = "Net Annual Yield (£)"

series = Series(
    yvalues,
    xvalues,
    title_from_data=False
)

scatter.series.append(series)

dashboard.add_chart(
    scatter,
    "H20"
)

# ==========================
# SAVE
# ==========================

wb.save(
    "NestMetrics_Analysis.xlsx"
)

print(
    "NestMetrics_Analysis.xlsx created successfully."
)
```
